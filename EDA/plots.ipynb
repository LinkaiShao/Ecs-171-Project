```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
student_mat = pd.read_csv('student-mat.csv', sep=';')
student_por = pd.read_csv('student-por.csv', sep=';')

# Combine the datasets
student = pd.concat([student_mat, student_por], ignore_index=True)

# One-hot encoding for categorical features (keep all categories)
student_encoded = pd.get_dummies(student, drop_first=False)

# Calculate correlation matrix for all columns
correlation_matrix = student_encoded.corr()

# Heatmap to visualize correlations between all features, including categorical features
plt.figure(figsize=(25, 20))
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm', linewidths=0.1)
plt.title('Correlation Heatmap of Student Dataset (Including All Encoded Features)')
plt.show()

# Selecting only numeric columns for correlation
numeric_cols = student.select_dtypes(include='number')

# Heatmap to visualize correlations between numeric features
plt.figure(figsize=(12, 8))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Student Dataset')
plt.show()

# Box plot for final grade (G3) distribution by school
plt.figure(figsize=(10, 6))
sns.boxplot(x='school', y='G3', data=student)
plt.title('Final Grade (G3) Distribution by School')
plt.xlabel('School')
plt.ylabel('Final Grade (G3)')
plt.show()

# Histogram for final grade (G3)
plt.figure(figsize=(10, 6))
sns.histplot(student['G3'], kde=True, bins=20)
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Frequency')
plt.show()


```
